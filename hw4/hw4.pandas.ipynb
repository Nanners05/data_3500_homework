```python
import pandas as pd
df = pd.read_csv("/workspaces/data_3500_homework/TSLA.csv")
```

```python
df["Close/Last"] = df["Close/Last"].replace("[$]", "", regex=True).astype(float)
df.head(10)
```

```python
df = df.iloc[::-1].reset_index(drop=True)
df.head(10)
```

```python
df["Moving/Avg"] = df["Close/Last"].shift(1).rolling(window=5).mean()
df.head(10)
```

```python
buy = 0
first_buy = 0
total_profit = 0

for i in range(len(df)):
    current_price = df.loc[i, "Close/Last"]
    avg_price = df.loc[i, "Moving/Avg"]

    if current_price < (avg_price * .98) and buy == 0:
        buy = current_price
        print(buy)
        if first_buy == 0:
            first_buy = current_price

    if current_price > (avg_price * 1.02) and buy != 0:
        trade_profit = current_price - buy
        buy = 0
        print(trade_profit)
        total_profit += trade_profit
```
